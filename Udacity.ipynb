```python
import requests

#Fetch a webpage
read = requests.get('https://news.ycombinator.com/')
print(read.text)
```

```python
import re

#Remove HTML Tags using RegEx
pattern = re.compile(r"<.*?>")  # Tags look like <....>
print(pattern.sub('',read.text))   #Replace them with blank
```

```python
from bs4 import BeautifulSoup

#Removing HTML tags using Beautiful Soup Library
soup =  BeautifulSoup(read.text,"html5lib")
print(soup.get_text())
```

```python
# Find all course summaries

sumi = soup.find_all("td", class_="storylink")
sumi[0]
```
